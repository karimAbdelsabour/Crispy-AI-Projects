```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
salary=pd.read_csv('/content/Salary.csv')
salary
```

```python
x=salary.iloc[:,0]
y=salary.iloc[:,1]
```

```python
sal_model=Sequential()
sal_model.add(Dense(1,activation='linear',input_dim=1))
```

```python
sal_model.summary()
```

```python
sal_model.compile(loss='mse', optimizer ='sgd')
```

```python
var=sal_model.fit(x,y,epochs=500,verbose=2) #verbose:It only affects the output apperance
```

```python
y_predict=sal_model.predict(x)
y_predict
```

```python
var.history['loss'][0]
```

```python
plt.plot((var.history['loss']),label='loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(loc='lower right')
plt.show
```
