```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
```

```python
data ={'outlooks':['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny',

'Rainy','Sunny','Overcast','Overcast','Rainy'],

'temp':['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild'],

'Humidity':['High','High','High','High','Normal','Normal','Normal','High','Normal','Normal','Normal','High','Normal','High'],



'Windy':['False','True','False','False','False','True','True','False','False','False','True','True','False','True'],





'play':['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']}
```

```python
df=pd.DataFrame(data)
df
```

```python
df["temp"]=df["temp"].astype('category')
df["temp"]=df["temp"].cat.codes
df["outlooks"]=df["outlooks"].astype('category')
df["outlooks"]=df["outlooks"].cat.codes
df["Humidity"]=df["Humidity"].astype('category')
df["Humidity"]=df["Humidity"].cat.codes
df["Windy"]=df["Windy"].astype('category')
df["Windy"]=df["Windy"].cat.codes
df["play"]=df["play"].astype('category')
df["play"]=df["play"].cat.codes
df
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
(x,y)
```

```python
model = GaussianNB()
model.fit(x,y)
y_predict = model.predict(x)

```

```python
accuracy=accuracy_score(y,y_predict)
accuracy
```
